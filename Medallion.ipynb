```python
import pandas as pd

def safe_not_equal(a, b):
    if pd.isna(a) and pd.isna(b):
        return False
    if pd.isna(a) or pd.isna(b):
        return True
    return a != b

df_sorted = df_append.sort_values(by=['idCredito', 'fechaEnvioCobro'])

cambios_list = []

for credito_id, grupo in df_sorted.groupby('idCredito'):
    grupo = grupo.reset_index(drop=True)
    idx_exito = grupo[grupo['isSuccessful'] == True].first_valid_index()
    
    if idx_exito is not None and idx_exito > 0:
        prev = grupo.iloc[idx_exito - 1]
        actual = grupo.iloc[idx_exito]
        
        for col in grupo.columns:
            if col in ['isSuccessful', 'idCredito']:
                continue
            
            if pd.isna(prev[col]) and pd.notna(actual[col]):
                cambios_list.append({
                    'idCredito': credito_id,
                    'columna': col,
                    'valor_anterior': None,
                    'valor_nuevo': actual[col]
                })
            elif safe_not_equal(prev[col], actual[col]):
                cambios_list.append({
                    'idCredito': credito_id,
                    'columna': col,
                    'valor_anterior': prev[col],
                    'valor_nuevo': actual[col]
                })

df_cambios = pd.DataFrame(cambios_list)

pd.set_option('display.max_colwidth', None)
display(df_cambios)
```
